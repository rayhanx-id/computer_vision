```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

im3 = cv2.imread("buku2.jpeg")
im = cv2.cvtColor(im3, cv2.COLOR_BGR2GRAY)
cv2_imshow(im)

mag_im = cv2.Canny(im, 100, 200)
cv2_imshow(mag_im)

TH = 150
lines = cv2.HoughLines(mag_im, 1, np.pi/180, TH)
res = im3.copy()

for r_t in lines:
    rho = r_t[0, 0]
    theta = r_t[0, 1]

    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho
    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))

    res = cv2.line(res, (x1, y1), (x2, y2), (255, 0, 0), thickness=2)

cv2_imshow(res)
```
