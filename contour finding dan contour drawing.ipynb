```python
import cv2 #import library OpenCV, (kalau di P. citra pakai PIL, di com vision pakai OpenCV)
from google.colab.patches import cv2_imshow #(google colab tidak support imshow, jadi dia bikin cara in buat menggantikan)
  
# buka gambar nya
image = cv2.imread('bangundatar.png')
  
# Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Canny edges detection (akan didapatkan di pengolahan citra pert akhir2)
edged = cv2.Canny(gray, 20, 200)

#tampilkan citra hasil edge detect
cv2_imshow(edged)
  
# Find Contours
contours = cv2.findContours(edged, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)[0]

#gambar dan print
# -1 menandakan menggambar semua contour
print("jumlah bangun yang ditemukan = " + str(len(contours)))
cv2.drawContours(image, contours, -1, (0, 200, 0), 3)
cv2_imshow(image)

```
